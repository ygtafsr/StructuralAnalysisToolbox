```python
import structuralanalysistoolbox as stbox
```

### Create Model Object

```python
model = stbox.Model(name="my-model")
```

### Import Mesh

```python
model.import_mesh(r'C:\Users\yigit\StructuralAnalysisToolbox\examples\meshes\cantilevier_beam.cdb')
model.plot_mesh()
```

### Select and Assign Material

```python
stbox.matlib.list_materials()
```

```python
steel = model.add_material(mat="Linear Steel")
model.assign_material("MESH", steel)
```

```python
model.info()
```

### Create Load Steps

```python
load_step_1 = model.add_loadstep(name="Pressure_Load_1")
```

Apply Pressure Through Surface Elements (SURF154+SF)

```python
load_step_1.pressure(nset="LOADING_NODES",
                     value= 20.0,
                     applied_by="SURFACE ELEMENTS",
                     operation="NEW")
load_step_1.dof("NS_FIX", "ALL", 0)

load_step_1.output("NODAL DOF", "ALL")
load_step_1.output("REACTION LOADS", "ALL")
load_step_1.output("NODAL-AVG PLASTIC STRAINS", "ALL")
load_step_1.output("NODAL-AVG STRESSES", "ALL")
load_step_1.output("EL-NODAL LOAD", "ALL")
```
