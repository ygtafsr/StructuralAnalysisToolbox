Import stbox Library

```python
import structuralanalysistoolbox as stbox
```

Create Model Object

```python
model = stbox.Model(name="my-model")
```

Import Mesh into the 'model'

```python
model.import_mesh(r'C:\Users\yigit\StructuralAnalysisToolbox\examples\meshes\cantilevier_beam.cdb')
model.plot_mesh()
```

Select and Assign Material

```python
# List pre-defined stbox library materials
stbox.matlib.list_materials()
```

```python
# Select and Assign Material
steel = model.add_material(mat="Linear Steel")
model.assign_material("MESH", steel)
```

Create Load Steps

```python
load_step_1 = model.add_loadstep(name="Pressure_Load_1")
```

Apply Boundary Conditions

```python
# Apply Force Through Element Faces (SFE)
load_step_1.force(nset="LOADING_NODES",
                  value= -1200,
                  direction="Y",
                  applied_by="ELEMENT FACES",
                  operation="NEW")

# Apply dof constraints
load_step_1.dof("NS_FIX", "ALL", 0)
```

Specify Load Step Outputs

```python
load_step_1.output("NODAL DOF", "ALL")
load_step_1.output("REACTION LOADS", "ALL")
load_step_1.output("NODAL-AVG PLASTIC STRAINS", "ALL")
load_step_1.output("NODAL-AVG STRESSES", "ALL")
load_step_1.output("EL-NODAL LOAD", "ALL")
```

Check Model Configuration

```python
model.info()
```

Solve Model

```python
model.solve()
#model.write_input_file()
```

```python
result = model.solution()
```

```python
# Displacement Result
disp = result.displacement()
disp.animate(deform=True, factor=1.0)
```
